```python
#LAUNCH YOUR PROJECT

project = ("Calender", 1.3, 2026, "Python", "Naqib")
print("=" * 40)
print(f"Project Name : {project[0]}")
print(f"Version : {project[1]}")
print(f"Started : {project[2]}")
print(f"Language : {project[3]}")
print(f"Lead name : {project[4]}")
print("=" * 40)
print(project[0:3])
print(f"Language count: {project.count("Python")}")
print(f"Language index :{project.count("Python")}")
print("=" * 40)

#tuples are immutable so python will raise type error, this is good for this data as this data are not something that we want to change

contributors = []

for i in range(4):
    while True:
      name = input("name? : ")
      if name.isalpha():
        break
      else:
        print("Name must only be letters!")

    while True:
      role = input("role? : ")
      if role.isalpha():
        break
      else:
        print("Role must only be letters!")

    while True:
      language = input("language? : ")
      if language.isalpha():
        break
      else:
        print("Language must only be letters!")

    commits = input("commits? : ")

    while True:
      country = input("country? : ")
      if country.isalpha():
        break
      else:
        print("Country must only be letters!")

    temp_dict = {"name": name, "role": role, "language": language, "commits": commits, "country": country}
    contributors.append(temp_dict)

    print("=" * 40)

for i, contributor_dict in enumerate(contributors):
    contributors[i].update({"status": "Active"})

# Access the status of the first contributor in the list
print(contributors[0].get("status"))

contributors_copy = contributors.copy()
print(f"Backup : {contributors_copy}")
```

```python
#TRACK AND ANALYSE ISSUES

issues = []

for i in range(5):
    id = input("id? : ")

    while True:
      title = input("title? : ")
      if title.isalpha():
        break
      else:
        print("Title must only be letters!")


    while True:
      issue_type = input("type? (Bug/Feature) : ")
      if issue_type == "Bug" or issue_type == "Feature":
        break
      else:
        print("Type must only be Bug / Feature!")

    while True:
      priority = input("priority? (Critical/High/Medium/Low) : ")
      if priority == "Critical" or priority == "High" or priority == "Medium" or priority == "Low":
        break
      else:
        print("Priority must only be Critical / High / Medium / Low!")

    while True:
      reporter = input("reporter? : ")
      if reporter.isalpha():
        break
      else:
        print("Reporter must only be letters!")

    while True:
      status = input("status? (Open/In Progress/Resolved) : ")
      if status == "Open" or status == "In Progress" or status == "Resolved":
        break
      else:
        print("Status must only be Open / In Progress / Resolved!")

    temp_dict = {"id": id, "title": title, "type": issue_type, "priority": priority, "reporter": reporter, "status": status}
    issues.append(temp_dict)

    print("=" * 40)

counter = 0

for issue_dict in issues:
    if issue_dict["status"] == "Open":
      counter += 1

print(f"Open issues : {counter}")

issues[0]["priority"] == "Critical"

print(f"First issue priority : {issues[0]["priority"]}")

print(f"Last two issues : {issues[3]}, {issues[4]}")

print("=" * 40)

reporters = set()

for issue_dict in issues:
  reporters.add(issue_dict["reporter"])

print(f"Reporters : {reporters}")

tech_stack = set()

for contributor_dict in contributors:
  tech_stack.add(contributor_dict["language"])

print(f"Tech stack : {tech_stack}")

tech_stack.add("Bingo")
tech_stack.remove("Bingo")

print(f"union : {reporters.union(tech_stack)}")
print(f"intersection : {reporters.intersection(tech_stack)}")
print(f"difference : {reporters.difference(tech_stack)}")

priorities_set = set()

for issue_dict in issues:
    priorities_set.add(issue_dict["priority"])

if "Critical" in priorities_set:
    print("'Critical' priority is present in the issues.")
else:
    print("'Critical' priority is not present in the issues.")

priority_count = {
    "Critical": 0,
    "High": 0,
    "Medium": 0,
    "Low": 0
}

for issue_dict in issues:
    priority = issue_dict["priority"]
    priority_count[priority] = priority_count.get(priority, 0) + 1

print(f"Priority keys: {priority_count.keys()}")
print(f"Priority values: {priority_count.values()}")

status_groups = {
    "Open": [],
    "In Progress": [],
    "Resolved": []
}

for issue_dict in issues:
    status = issue_dict["status"]
    title = issue_dict["title"]
    status_groups[status].append(title)

print(f"Status items: {status_groups.items()}")

reporter_counts = {}

for issue_dict in issues:
    reporter = issue_dict["reporter"]
    reporter_counts[reporter] = reporter_counts.get(reporter, 0) + 1

top_reporter = None
max_count = 0

for reporter, count in reporter_counts.items():
    if count > max_count:
        max_count = count
        top_reporter = reporter

print(f"The top reporter is: {top_reporter} with {max_count} issues.")

first_issue = issues[0]
removed_type = first_issue.pop('type')

print(f"Removed 'type': {removed_type}")
print(f"First issue after removing 'type': {first_issue}")
```

```python
#SAVE, READ AND REPORT

import os

folder_name = project[0].lower().replace(' ', '_')
project_folder = folder_name

if not os.path.exists(project_folder):
    os.makedirs(project_folder)
    print(f"Created folder: {project_folder}")
else:
    print(f"Folder already exists: {project_folder}")

print(f"Folder contents of '{project_folder}': {os.listdir(project_folder)}")
print("=" * 40)

report_file_path = os.path.join(project_folder, "project_report.txt")

report_content = f"""
========================================
PROJECT REPORT
========================================
Project Name: {project[0]}
Version: {project[1]}
Started: {project[2]}
Language: {project[3]}
Lead Name: {project[4]}

Contributors:
"""
for contributor in contributors:
    report_content += f"- Name: {contributor['name']}, Role: {contributor['role']}, Language: {contributor['language']}, Commits: {contributor['commits']}, Country: {contributor['country']}, Status: {contributor['status']}\n"

report_content += f"""

Issues:
"""
for issue in issues:
    issue_type = issue.get('type', '')
    report_content += f"- ID: {issue['id']}, Title: {issue['title']}, Type: {issue_type}, Priority: {issue['priority']}, Reporter: {issue['reporter']}, Status: {issue['status']}\n"

report_content += f"""

Analysis:
----------------------------------------
Open Issues: {counter}
Reporters: {reporters}
Tech Stack: {tech_stack}
Union (Reporters and Tech Stack): {reporters.union(tech_stack)}
Intersection (Reporters and Tech Stack): {reporters.intersection(tech_stack)}
Difference (Reporters not in Tech Stack): {reporters.difference(tech_stack)}
Critical Priority Present: {'Critical' in priorities_set}
Priority Breakdown: {priority_count}
Status Groups: {status_groups}
Top Reporter: {top_reporter} with {max_count} issues.
"""

try:
    with open(report_file_path, 'w') as f:
        f.write(report_content)
    print(f"Project report saved to: {report_file_path}")
except IOError as e:
    print(f"Error saving project report: {e}")
print("=" * 40)

issues_csv_file_path = os.path.join(project_folder, "issues.csv")

csv_header = "id,title,priority,reporter,status\n"
csv_rows = []
for issue in issues:
    csv_rows.append(f"{issue['id']},{issue['title']},{issue['priority']},{issue['reporter']},{issue['status']}\n")

try:
    with open(issues_csv_file_path, 'w') as f:
        f.write(csv_header)
        f.writelines(csv_rows)
    print(f"Issues data saved to: {issues_csv_file_path}")
except IOError as e:
    print(f"Error saving issues CSV: {e}")
print("=" * 40)

print(f"Reading back from {report_file_path}:")

try:
    with open(report_file_path, 'r') as f:
        print("\n--- Full file content (read()) ---")
        print(f.read())

    with open(report_file_path, 'r') as f:
        print("\n--- First two lines (readline()) ---")
        print(f.readline(), end='')
        print(f.readline(), end='')

    with open(report_file_path, 'r') as f:
        all_lines = f.readlines()
        print(f"\n--- readlines() ---")
        print(f"Total lines : {len(all_lines)}")
        counter = 0
        for line in all_lines:
            if 'Critical' in line or 'High' in line:
                counter += 1
        print(f"Total lines containing 'Critical' or 'High': {counter}\n")

except FileNotFoundError:
    print(f"Error: File not found at {report_file_path}")
print("=" * 40)

contributor_names = sorted([c['name'] for c in contributors])

print(f"FINAL SUMMARY")
print("=" * 40)
print(f"Project: {project[0]} (Version {project[1]})")
print(f"Contributors: {', '.join(contributor_names)}")
print(f"Total Issues: {len(issues)}")
print(f"Open Issues: {counter}")
print(f"Reporters: {reporters}")
print(f"Tech Stack: {tech_stack}")
print(f"Top Reporter: {top_reporter} with {max_count} issues.")
print(f"Priority Breakdown: {priority_count}")
print(f"Report file path: {report_file_path}")
print(f"Issues CSV file path: {issues_csv_file_path}")
```
